```python
# get a list of sp500 companies and their sector
# from wikipedia

import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

# get the list of sp500 companies and their sector

url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
table = soup.find('table', {'class': 'wikitable sortable'})
rows = table.find_all('tr')
sp500 = []
for row in rows[1:]:
    cols = row.find_all('td')
    sp500.append([cols[0].text.strip(), cols[2].text.strip()])

```

```python
#show the table
print(sp500)
#make a dataframe
df = pd.DataFrame(sp500, columns=['Symbol', 'Sector'])
sp500
```

```python
import pandas as pd
import numpy as np

df_info_tech = df[df['Sector'] == 'Information Technology']['Symbol']
symbols_list = df_info_tech.values.T.tolist()
```

```python
import yfinance as yf

dates = [['2018-01-01', '2019-01-01'], ['2019-01-01', '2020-01-01'], ['2020-01-01', '2021-01-01'], ['2021-01-01', '2022-01-01'], ['2022-01-01', '2023-01-01'], ['2023-01-01', '2024-01-01']]
for date in dates:
    data = [yf.Tickers(symbols_list).history(period="1d", start=date[0], end=date[1])]
```
